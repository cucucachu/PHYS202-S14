<Strong>(a)</Strong> Use the numerical solution to see what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs. t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 or π ?

```python
from scipy.integrate import odeint;

def pendulum(th0, w0, length, g, t = np.arange(0.,10.+0.01,0.01)):

    def deriv(theta_current,t):  # return derivatives of the array theta
        theta = theta_current[0]
        omega = theta_current[1]
        dtheta = omega
        domega = -(g/length)*np.sin(theta)
        return np.array([ dtheta, domega ])    
    
    theta_init = np.array([th0, w0]) # initial values, th0 from before = 20 degrees, expressed in radians

    #pass the function deriv to the solver with initial values and integration variable
    theta = odeint(deriv,theta_init,t);
    
    return theta;
```

```python
t = np.arange(0.,10.+0.01,.01);
length = 1;
angles = np.arange(0, np.pi + np.pi/4, np.pi/4);

for angle in angles:
    label = "theta_0: %.2f\pi" % (angle/np.pi);
    plt.plot(t, pendulum(angle, 0,length,9.8)[:,0], label=label);

plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-3,3)
plt.grid()
plt.legend()
plt.show();
```

As $\theta_0$ increases from 0 to $\pi$, the pendulum's amplitude increases, and its frequency decreases. When $\theta_0 = 0$ or $\theta_0 = \pi$, the pendulum will not swing.

## Question 2

**(a)** Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
$\frac{dy}{dx} = x + 2y$
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
dydx = lambda y,x: x + 2*y;

y = 0;
dx = .25;
y1 = [];
x1 = np.arange(0, 1.+dx, dx);
for x in x1:
    y = y + dydx(y,x)*dx;
    y1.append(y);
    
print "Euler method approximation with dx=.25:", y
    
y = 0;
dx = .02;
y2 = [];
x2 = np.arange(0, 1.+dx, dx);
for x in x2:
    y = y + dydx(y,x)*dx;
    y2.append(y);

print "Euler method approximation with dx=.02:", y
```

**(b)** Plot the results from part (a) for the two different step sizes with the exact solution
$y=0.25e^{2x}−0.5x−0.25$
on one plot. Include a legend. Which step size gives a more accurate result?

```python
x = np.arange(0, 1.+dx, dx);
y = .25*np.exp(2*x) - 0.5*x -0.25;

y1 = np.array(y1);
y2 = np.array(y2);

plt.figure(figsize=(10,8));
plt.plot(x, y, "go", x, y, "g-", linewidth=2., label="Exact Solution");
plt.plot(x1, y1, "ro", x1, y1, "r-", label="Eueler (dx=.25)");
plt.plot(x2, y2, "yo", x2, y2, "y-", label="Eueler (dx=.02)");
plt.legend();
plt.xlabel("x", size=10.0);
plt.ylabel("$y(x) = .25e^{x^{2}} - 0.5x - 0.25$", size=20);
plt.plot();
```

**(c)** Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
dydx = lambda y,x: x + 2*y;

y = 0;
dx = .25;
y1_cromer = [];
x1 = np.arange(0, 1.+dx, dx);
for x in x1:
    y = y + dydx(y,x+dx)*dx;
    y1_cromer.append(y);
    
print "Euler/Cromer method approximation with dx=.25:", y
    
y = 0;
dx = .02;
y2_cromer = [];
x2 = np.arange(0, 1.+dx, dx);
for x in x2:
    y = y + dydx(y,x)*dx;
    y2_cromer.append(y);

print "Euler/Cromer method method approximation with dx=.02:", y

y = 0;
dx = .25;
y1_midpoint = [];
x1 = np.arange(0, 1.+dx, dx);
for x in x1:
    y = y + dydx(y,x)*dx;
    y1_midpoint.append(y);
    
print "Midpoint method approximation with dx=.25:", y
    
y = 0;
dx = .02;
y2_midpoint = [];
x2 = np.arange(0, 1.+dx, dx);
for x in x2:
    y = y + dydx(y,x)*dx;
    y2_midpoint.append(y);

print "Midpoint method approximation with dx=.02:", y
print "y(1) =", .25*np.exp(2) - .5 - .25;
```

```python
x = np.arange(0, 1.+dx, dx);
y = .25*np.exp(2*x) - 0.5*x -0.25;

y1 = np.array(y1);
y2 = np.array(y2);

y1_cromer = np.array(y1_cromer);
y2_cromer = np.array(y2_cromer);

y1_midpoint = np.array(y1_midpoint);
y2_midpoint = np.array(y2_midpoint);

plt.figure(figsize=(15,20));
#3f, (plt1, plt2, plt3, plt4) = plt.subplots(4, sharex=True, sharey=True)

plt.subplot(221);
plt.title("Euler Method Approx.", size=20)
plt.plot(x, y, "go", x, y, "g-", linewidth=2., label="Exact Solution");
plt.plot(x1, y1, "ro", x1, y1, "r-", label="Eueler (dx=.25)");
plt.plot(x2, y2, "yo", x2, y2, "y-", label="Eueler (dx=.02)");
plt.legend();
plt.xlabel("x", size=10.0);
plt.ylabel("$y(x) = .25e^{x^{2}} - 0.5x - 0.25$", size=20);


plt.subplot(222);
plt.title("Euler/Cromer Approx.", size = 20)
plt.plot(x, y, "go", x, y, "g-", linewidth=2., label="Exact Solution");
plt.plot(x1, y1_cromer, "ro", x1, y1_cromer, "r-", label="Euler/Cromer (dx=.25)");
plt.plot(x2, y2_cromer, "yo", x2, y2_cromer, "y-", label="Euler/Cromer (dx=.02)");
plt.legend();
plt.xlabel("x", size=10.0);
plt.ylabel("$y(x) = .25e^{x^{2}} - 0.5x - 0.25$", size=20);



plt.subplot(223);
plt.title("Midpoint Approximation", size = 20);
plt.plot(x, y, "go", x, y, "g-", linewidth=2., label="Exact Solution");
plt.plot(x1, y1_midpoint, "ro", x1, y1_midpoint, "r-", label="Midpoint (dx=.25)");
plt.plot(x2, y2_midpoint, "yo", x2, y2_midpoint, "y-", label="Midpoint (dx=.02)");
plt.legend();
plt.xlabel("x", size=10.0);
plt.ylabel("$y(x) = .25e^{x^{2}} - 0.5x - 0.25$", size=20);


plt.subplot(224);
plt.title("Percent Error", size = 20)
plt.plot(x2[1:], 100*np.abs(y2[1:]-y[1:])/y[1:], label="Euler %error");
plt.plot(x2[1:], 100*np.abs(y2_cromer[1:]-y[1:])/y[1:], label="Euler/Cromer %error");
plt.plot(x2[1:], 100*np.abs(y2_midpoint[1:]-y[1:])/y[1:], label="Midpoint %error");
plt.xlabel("x");
plt.ylabel("% error", size=15)
plt.legend();

error = lambda calc, actual: np.nan if actual == 0 else 100*np.abs(calc-actual)/actual;
y1_exact = .25*np.exp(2) - .5 - .25;

errors = [error(y2[-1], y1_exact),
          error(y2_cromer[-1], y1_exact),
          error(y2_midpoint[-1], y1_exact) ];

if errors[2] == min(errors):
    print "The midpoint approximation with dx = .02 is the best approximation of the three"
```

We can use numerical solvers and widgets to explore the dynamics of a projectile that is subject to a drag force proportional to the square of its velocity - "quadratic drag".
The force on the projectile is given by
<p style="margin: auto">
$\sum\vec{F} = m\vec{a}\\ = m\vec{g} - c v^2\hat{v}\\ = m\vec{g} - c v \vec{v}\\ = m\vec{g} - c \sqrt{v_x^2 + v_y^2}\vec{v}$
</p>
Where the second term is the drag force, in the direction opposite that of the velocity vector. Re-writing these in terms of the first-order differential equations for vx and vy gives

$m\frac{dv_x}{dt} = -c \sqrt{v_x^2 + v_y^2} v_x$ <br>
$m\frac{dv_y}{dt} = mg -c \sqrt{v_x^2 + v_y^2} v_y$

taking the positive y direction to be down. Notice that these are coupled equations. There are both vx and vy terms in each one. The drag coefficient c depends on the shape and density of the projectile. We can re-write this system of equations in terms of the terminal velocity - the point at which the gravitational force balances the drag force. This happens when $mg=cv^{2}$. Let's denote the terminal velocity as $v_{ter} = \sqrt{mg/c}$ and re-write our equations in terms of $v_ter$.

$\frac{dv_x}{dt} = \frac{-g}{m} \frac{v_x}{v_{ter}^2} \sqrt{v_x^2 + v_y^2}$<br>
$\frac{dv_y}{dt} = \frac{-g}{m}\left(1 + \frac{v_y}{v_{ter}^2} \sqrt{v_x^2 + v_y^2}\right)$

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.
