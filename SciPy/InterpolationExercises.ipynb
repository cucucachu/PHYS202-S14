<p><strong>(a)</strong> Let's turn this look-up table into a continuous function. Write a reusable function <code>Temp(Resistance)</code> that takes any value of <code>Resistance</code> and returns the interpolated temperature. Be sure to include a docstring in your function.</p>

```python
from scipy.interpolate import interp1d

def Temp(Resistance):
    """Returns the tempurature of the device at a given resistance using
        a linear interpolation"""
    temp = arange(10, 40)
    res = [3.239, 3.118, 3.004, 2.897, 2.795, 2.700, 2.610, 2.526, 2.446, 2.371, 2.300, 2.233, 2.169, 2.110, 2.053, 2.000, 1.950, 1.902, 1.857, 1.815, 1.774, 1.736, 1.700, 1.666, 1.634, 1.603, 1.574, 1.547, 1.521, 1.496];
    res = np.array(res)
    res = res[::-1];
    temp = temp[::-1];
    TvR = interp1d(res, temp)
    return TvR(Resistance)
```

<p><strong>(b)</strong> Show that the function returns expected values, including some from the table for a few specific input resistances.</p>

```python
print "Temp(3) should return about 12";
print "   My function returns", str(Temp(3));
print "Temp(2.700) should return 15";
print "   My function returns", str(Temp(2.7));
print "Temp(1.5) should return 39";
print "   My function returns", str(Temp(1.5));
```
