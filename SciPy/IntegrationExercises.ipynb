```python
%%file Integrators.py

import numpy as np;

def trapz(func, a, b, N):
    """Returns the numerical integration of the function func, from the limits a to b.
    This function uses the trapazoidal method to approximate the integral. It will use
    N slices, and becomes more accurate and more computationaly expensive as N increases.
    """
    h = (b - a) / N;
    k = np.arange(1, N);
    return h*(0.5*func(a) + 0.5*func(b) + func(a+k*h).sum());

def simps(func, a, b, N):
    """Returns the numerical integration of the function func, from the limits a to b.
    This function uses Simpson's method to approximate the integral. It will use
    N slices, and becomes more accurate and more computationaly expensive as N increases.
    """
    h = (b - a) / N;
    k1 = np.arange(1, N / 2 + 1);
    k2 = np.arange(1, N / 2);
    return (1./3.)*h*(func(a) + func(b) + 4.*func(a+(2*k1-1)*h).sum() + 2.*func(a+2*k2*h).sum());
```

```python
from scipy.integrate import quad
from scipy.integrate import trapz

x = np.linspace(1./3., .5, 100)

def Psi1(x):
    return (np.sqrt(2) * np.sin(np.pi * x)) ** 2;

def Psi2(x):
    return (np.sqrt(2) * np.sin(2 * np.pi * x)) ** 2;

psi1Arr = Psi1(x);
psi2Arr = Psi2(x);

print "Probability of finding the particle between 1/3 and 1/2 \
for the particle in ground state."
print "   Using trapz:", trapz(psi1Arr, x);
print "   Using quad: %f" % quad(Psi1, 1/3., 1/2., )[0];
print "Probability of finding the particle between 1/3 and 1/2 \
for the particle in first excited state."
print "   Using trapz:", trapz(psi2Arr, x);
print "   Using quad: ", quad(Psi2, 1/3., 1/2., );
print "\n100 slices is enough for trapz to be accurate to 6 decimal places."
```
