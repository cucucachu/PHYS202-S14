```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint

C = -2
timeLimit = 99

loc = [-1, 10]

def R(r, t):
    accel = C / (r[1]**2);
    vel = r[0];
    return np.array((accel, vel))

loc = odeint(R, loc, np.linspace(0, timeLimit, timeLimit + 1));

for i in range(len(loc)):
    print 'time: %2.2f   speed: %2.3f   r: %2.3f' % (i, loc[i, 0], loc[i, 1])
    
#plt.plot(time, loc);
#plt.show();
```
