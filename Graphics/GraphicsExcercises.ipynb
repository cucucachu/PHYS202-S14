<p><strong>(b)</strong> Now add code to calculate the running average of the data. Plot both the original data and the running average on the same graph, only over the range covered by the first 1000 data points.</p>

```python
def runningAverage(data):
    avg = [];
    for num in range(1, len(data) + 1):
        sum = 0;
        for i in range(num):
            sum = sum + data[i];
        avg.append(float(sum) / num);
    return avg;

months = np.loadtxt('sunspots.txt', dtype='int', usecols = (0,));
sunspots = np.loadtxt('sunspots.txt', dtype='float', usecols=(1,));
years = 1749 + months / 12;
years = years[::12];

plt.figure(figsize=(15, 12), dpi=100)
plt.plot(months, sunspots, color='y', linewidth='2', marker='*', markersize=4, label='# of sunspots')
plt.plot(months, runningAverage(sunspots), color='b', linewidth='2', label='Running Average')
plt.legend(loc='upper left');
plt.suptitle('Sunspots from 1750 to 2010', fontsize=35);
plt.xlabel('Year',fontsize=20)
plt.ylabel('# of Sunspots',fontsize=20)
plt.xticks(np.arange(0, 3500, 12* 25), np.arange(1750, 2010, 25))

for label in plt.gca().get_xticklabels() + plt.gca().get_yticklabels():
    label.set_fontsize(15)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))

    plt.grid();

plt.show()
```
